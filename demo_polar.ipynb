```python
# Import the POLAR package
import numpy as np
import polar
```

```python

# Generate a random signal
t = np.arange(100)*.01 # time vector
x = np.linspace(0,2,50) # spatial coordinates

s = polar.signal.Signal(np.random.randn(t.size,x.size),
                        dims=('t','x'),
                        coords=[("t",t),("x",x)]
                        )

s.plot()

F = s.fft('t')
S = F.ifft('f')
```
